```python
def JdToStD (jdate):
    fmt = '%Y%j'
    datestd = datetime.datetime.strptime(jdate, fmt).date()
    return(datestd)

def convData(data):
    data['Date'] = pd.DataFrame(np.zeros((len(data),1)))
    data['Day'] = pd.DataFrame(np.zeros((len(data),1)))
    data['Month'] = pd.DataFrame(np.zeros((len(data),1)))

    for i in range(len(data)):
        s = str(round(data['Year_LBAMIP'].loc[i]))+str(round(data['DoY_LBAMIP'].loc[i]))
        data['Date'].loc[i] = JdToStD(s)
        data['Month'].loc[i] = datetime.datetime.strptime(s,'%Y%j').month
        data['Day'].loc[i] = datetime.datetime.strptime(s,'%Y%j').day

    data['Date'] = pd.to_datetime(data['Date'].values)
```

```python
def scaling(train, test):
    ss = StandardScaler()
    ss.fit(train[['ta', 'press', 'par', 'Rn', 'H', 'LE', 'ee', 'ees', 'So', 'prec','NEEnogap_5day_sco2_ust']])

    train_sc = pd.DataFrame(
        ss.transform(
            train[['ta', 'press', 'par', 'Rn', 'H', 'LE', 'ee', 'ees', 'So', 'prec','NEEnogap_5day_sco2_ust']]
        ),
        index = train['Date'].values,
        columns = train.columns[1:]
    ).reset_index().rename(columns={'index':'Date'})

    test_sc = pd.DataFrame(
        ss.transform(
            test[['ta', 'press', 'par', 'Rn', 'H', 'LE', 'ee', 'ees', 'So', 'prec','NEEnogap_5day_sco2_ust']]
        ),
        index = test['Date'].values,
        columns = test.columns[1:]
    ).reset_index().rename(columns={'index':'Date'})

    return ss, train_sc, test_sc
```

```python
# mice imputer


def imputer(train, test):
    mice_imputer = IterativeImputer(max_iter=10, random_state=0, sample_posterior=True)
    mice_imputer.fit(train[['ta', 'press', 'par', 'Rn', 'H', 'LE', 'ee', 'ees', 'So', 'prec', 'NEEnogap_5day_sco2_ust']])

    train_mice_imputed = pd.DataFrame(
        mice_imputer.transform(train[['ta', 'press', 'par', 'Rn', 'H', 'LE', 'ee', 'ees', 'So', 'prec', 'NEEnogap_5day_sco2_ust']]),
        index=train['Date'].values,
        columns=train.columns[1:]
    ).reset_index().rename(columns={'index': 'Date'})

    test_mice_imputed = pd.DataFrame(
        mice_imputer.transform(test[['ta', 'press', 'par', 'Rn', 'H', 'LE', 'ee', 'ees', 'So', 'prec', 'NEEnogap_5day_sco2_ust']]),
        index=test['Date'].values,
        columns=test.columns[1:]
    ).reset_index().rename(columns={'index': 'Date'})

    return  train_mice_imputed, test_mice_imputed
```

```python
def objective_rf(trial, X, y):

    X_train, X_val = X[:-round(X.shape[0]*0.2),:], X[-round(X.shape[0]*0.2):,:]
    y_train, y_val = y[:-round(y.shape[0]*0.2)], y[-round(y.shape[0]*0.2):]

    n_estimators = trial.suggest_int("n_estimators", 100, 2000)
    max_depth = trial.suggest_int("max_depth", 5, 20)
    min_samples_leaf = trial.suggest_int("min_samples_leaf", 50, 200)
    max_features = trial.suggest_float("max_features", 0.1, 1)

    rf = RandomForestRegressor(n_estimators=n_estimators, max_depth=max_depth, min_samples_leaf=min_samples_leaf, max_features=max_features, n_jobs=-1, random_state=10)
    rf.fit(X_train, y_train)

    pred = rf.predict(X_val)

    # Evaluate the model accuracy on the validation set.
    score = mean_squared_error(y_val, pred)
    return score
```

```python
X = train.drop(columns=['Date', 'tower', 'NEEnogap_5day_sco2_ust']).values
y = train['NEEnogap_5day_sco2_ust'].values

X_train, X_val = X[:-round(X.shape[0]*0.2),:], X[-round(X.shape[0]*0.2):,:]
y_train, y_val = y[:-round(y.shape[0]*0.2)], y[-round(y.shape[0]*0.2):]

input_shape = X_train.shape[1]

model = keras.Sequential()

model.add(layers.Dense(793, activation='relu', input_dim=X.shape[1]))
model.add(layers.Dropout(0.3035197246543597))
model.add(layers.BatchNormalization())

model.add(layers.Dense(743, activation='relu'))
model.add(layers.Dropout(0.3035197246543597))
model.add(layers.BatchNormalization())



model.add(layers.Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam', metrics = 'mse')

# We compile our model with a sampled learning rate.
epochs = 100000

earlystopper = EarlyStopping(monitor = 'val_loss', patience = 10, verbose = 1)

model.fit(
    X_train,
    y_train,
    batch_size=20,
    epochs=epochs,
    validation_data = (X_val, y_val),
    callbacks = [earlystopper],
    verbose=0)


# Evaluate the model accuracy on the validation set.
score_test = model.predict(test.drop(columns=['Date', 'tower', 'NEEnogap_5day_sco2_ust']).values)
score_train = model.predict(train.drop(columns=['Date', 'tower', 'NEEnogap_5day_sco2_ust']).values)

print(f'The r2 score on the Train set is: ', r2_score(train['NEEnogap_5day_sco2_ust'].values, score_train))
print(f'The r2 score on the Test set is ', r2_score(test['NEEnogap_5day_sco2_ust'].values, score_test))
```
